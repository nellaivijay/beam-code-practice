## Data Validation

```python
def validate_data(file_path, expected_columns=None):
    """Validate data file structure and content."""
    if not os.path.exists(file_path):
        return False, "File not found"
    
    try:
        if file_path.endswith('.csv'):
            df = pd.read_csv(file_path)
            if expected_columns:
                missing = set(expected_columns) - set(df.columns)
                if missing:
                    return False, f"Missing columns: {missing}"
        elif file_path.endswith('.json'):
            import json
            with open(file_path, 'r') as f:
                data = json.load(f)
        return True, "Validation successful"
    except Exception as e:
        return False, str(e)

expected_sales_columns = ['transaction_id', 'customer_id', 'product_id', 'quantity', 'price', 'timestamp']
valid, message = validate_data(sales_file, expected_sales_columns)
print(f"Sales data validation: {valid} - {message}")
```

## Data Directory Structure

```python
def show_directory_structure(path):
    """Display directory structure."""
    for root, dirs, files in os.walk(path):
        level = root.replace(path, '').count(os.sep)
        indent = ' ' * 2 * level
        print(f"{indent}{os.path.basename(root)}/")
        subindent = ' ' * 2 * (level + 1)
        for file in sorted(files):
            print(f"{subindent}{file}")

if os.path.exists(data_dir):
    print("Data directory structure:")
    show_directory_structure(data_dir)
else:
    print("Data directory not found.")
```

## Exercise Solutions

### Task 1: Generate Streaming Event Data

```python
import pandas as pd
import json
from datetime import datetime, timedelta
import random

# Generate streaming events
events = []
event_types = ['page_view', 'click', 'purchase', 'signup', 'logout']
base_time = datetime.now()

for i in range(1000):
    event = {
        'event_id': f"evt_{i:06d}",
        'user_id': f"user_{random.randint(1, 100):03d}",
        'event_type': random.choice(event_types),
        'timestamp': (base_time + timedelta(seconds=random.randint(0, 3600))).isoformat(),
        'properties': json.dumps({'page': f"/page/{random.randint(1, 10)}"})
    }
    events.append(event)

events_df = pd.DataFrame(events)
events_file = '/home/ramdov/projects/beam-code-practice/data/streaming_events.csv'
events_df.to_csv(events_file, index=False)
print(f"Generated {len(events)} streaming events")
print(events_df.head())
```

### Task 2: Generate Sensor Data

```python
# Generate sensor data
sensor_data = []
for i in range(500):
    reading = {
        'sensor_id': f"sensor_{random.randint(1, 20):02d}",
        'temperature': round(random.uniform(15.0, 35.0), 2),
        'humidity': round(random.uniform(30.0, 90.0), 2),
        'timestamp': (base_time + timedelta(minutes=random.randint(0, 1440))).isoformat(),
        'location': f"zone_{random.randint(1, 5)}"
    }
    sensor_data.append(reading)

sensor_df = pd.DataFrame(sensor_data)
sensor_file = '/home/ramdov/projects/beam-code-practice/data/sensor_readings.csv'
sensor_df.to_csv(sensor_file, index=False)
print(f"Generated {len(sensor_data)} sensor readings")
print(sensor_df.head())
```
